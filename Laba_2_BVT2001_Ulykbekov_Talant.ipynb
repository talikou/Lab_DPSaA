Алгоритмы поиска:

| [Бинарное дерево](https://en.wikipedia.org/wiki/Binary_tree) | 
| [Метод поиска Фибоначчи](https://en.wikipedia.org/wiki/Fibonacci_search_technique) | 
| [Интерполяционный поиск](https://en.wikipedia.org/wiki/Interpolation_search) |

```python
from math import floor
def binary_search(arr, elem):
    high = len(arr) - 1
    low = 0
    k = 0
    for i in arr:
        k += 1
        middle = floor((high+low)/2)
        if arr[middle] == elem:
            print(f"Кол-во попыток: {k}")
            return middle
        if arr[middle] > elem:
            high = middle - 1
        if arr[middle] < elem:
            low = middle - 1
```

```python

from random import randint
a = sorted([randint(-100, 100) for _ in range(100)])
print(a)
print(binary_search(a, 7))
```

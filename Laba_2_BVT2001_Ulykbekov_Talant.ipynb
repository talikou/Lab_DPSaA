```python
def add(arr, elem, index):
    if arr[index] is not None:
        print("this place is already filled")
        return
    arr[index] = elem
    return arr
```

```python
def delete(arr, index):
    if index >= len(arr):
        print("index out of range")
        return
    del arr[index]
    return arr
```

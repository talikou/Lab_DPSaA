### Выполнение:

```python
import turtle

def create_l_system(iters, axiom, rules):
    start_string = axiom
    if iters == 0:
        return axiom
    end_string = ""
    for _ in range(iters):
        end_string = "".join(rules[i] if i in rules else i for i in start_string)
        start_string = end_string

    return end_string


def draw_l_system(t, instructions, angle, distance):
    for cmd in instructions:
        if cmd == 'F':
            t.forward(distance)
        elif cmd == '+':
            t.right(angle)
        elif cmd == '-':
            t.left(angle)


def main(iterations, axiom, rules, angle, length=8, size=2, y_offset=0,
        x_offset=0, offset_angle=0, width=450, height=450):

    inst = create_l_system(iterations, axiom, rules)

    t = turtle.Turtle()
    wn = turtle.Screen()
    wn.setup(width, height)

    t.up()
    t.backward(-x_offset)
    t.left(90)
    t.backward(-y_offset)
    t.left(offset_angle)
    t.down()
    t.speed(0)
    t.pensize(size)
    draw_l_system(t, inst, angle, length)
    t.hideturtle()
    t.clear()
```

Кривая Коха

```python
def krivaya_koha():
    axiom = "F--F--F"
    rules = {"F":"F+F--F+F"}
    iterations = 4
    angle = 60
    main(iterations, axiom, rules, angle, x_offset=180, y_offset=-300, width=800, height=800)
```

Кривая Гильберта

```python
def krivaya_gilberta():
    axiom = "L"
    rules = {"L":"+RF-LFL-FR+", "R":"-LF+RFR+FL-"}
    iterations = 4
    angle = 90
    main(iterations, axiom, rules, angle, x_offset=-475, y_offset=-475, width=1000, height=1000)
```

Кривая Серпинского

```python
def krivaya_serpinskogo():
    axiom = "F+XF+F+XF"
    rules = {"X":"XF-F+F-XF+F+XF-F+F-X"}
    iterations = 4
    angle = 90
    main(iterations, axiom, rules, angle, x_offset=-235, width=600, height=600)
```

Салфетка Серпинского

```python
def salfetka_serpinskogo():
    axiom = "FXF--FF--FF"
    rules = {"F":"FF", "X":"--FXF++FXF++FXF--"}
    iterations = 5
    angle = 60
    main(iterations, axiom, rules, angle, x_offset=250, y_offset=-500, width=900, height=1200)
```

Ковер Серпинского

```python
def kover_serpinskogo():
    axiom = "YF"
    rules = {"X":"YF+XF+Y", "Y":"XF-YF-X"}
    iterations = 4
    angle = 60
    main(iterations, axiom, rules, angle, x_offset=-250, y_offset=-250, width=550, height=550)
```

Дерево Пифагора

```python
import turtle
import math

def derevo_pifagora():
    def fractal(aturt, depth, maxdepth):
        if depth > maxdepth:
            return
        length = 180 * ((math.sqrt(2) / 2) ** depth)
        anotherturt = aturt.clone()
        aturt.forward(length)
        aturt.left(45)
        fractal(aturt, depth + 1, maxdepth)
        anotherturt.right(90)
        anotherturt.forward(length)
        anotherturt.left(90)
        anotherturt.forward(length)
        if depth != maxdepth:
            turt3 = anotherturt.clone()
            turt3.left(45)
            turt3.forward(180 * ((math.sqrt(2) / 2) ** (1 + depth)))
            turt3.right(90)
            fractal(turt3, depth + 1, maxdepth)
        anotherturt.left(90)
        anotherturt.forward(length)


    def draw_fractal():
        window = turtle.Screen()
        window.setup(1200, 900)
        t = turtle.Turtle()
        t.hideturtle()
        t.penup()
        t.goto(-75, -225)
        t.pendown()
        t.speed(0)
        t.left(90)
        fractal(t, 1, 8)
    draw_fractal()
```

```python
import time
functions = {
    'Кривая Коха': krivaya_koha(),
    'Кривая Гильберта': krivaya_gilberta(),
    'Кривая Серпинского': krivaya_serpinskogo(),
    'Салфетка Серпинского': salfetka_serpinskogo(),
    'Ковер Серпинского': kover_serpinskogo(),
    'Дерево Пифагора': derevo_pifagora(),
}

for name, function in functions.items():
    time_start = time.perf_counter()
    print(f'{name} - {time.perf_counter() - time_start}')
```
